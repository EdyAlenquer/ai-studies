# Decision Tree


# Imports

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.append('../')

from utils.preprocessing import train_test_split, kfolds_cross_validation
from utils.output_utils import do_cv_and_get_metrics_classification
```

# Process

```python
data = np.genfromtxt('./data/kc2.csv', delimiter=',')
print('Shape:', data.shape)
data[:3, :]
X = data[:, :-1]
y = data[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 0.9, random_seed=466851
)

print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', X_test.shape)
print('y_test shape:', y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
```

## Gini Impurity

```python
decision_tree = DecisionTreeClassifier(criterion='gini')

cv_splits = kfolds_cross_validation(
    data=X_train,
    n_folds=10,
    shuffle=True
)

do_cv_and_get_metrics_classification(
    classifier=decision_tree, 
    cv_splits=cv_splits, 
    X_train=X_train, 
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    title=' Decision Tree - Gini Impurity '
)
```

```python
fig = plt.figure(figsize=(30,15))
tree.plot_tree(decision_tree, filled=True)
fig.show()
```

## Entropy

```python
decision_tree = DecisionTreeClassifier(criterion='entropy')

cv_splits = kfolds_cross_validation(
    data=X_train,
    n_folds=10,
    shuffle=True
)

do_cv_and_get_metrics_classification(
    classifier=decision_tree, 
    cv_splits=cv_splits, 
    X_train=X_train, 
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    title=' Decision Tree - Entropy '
)
```

```python
fig = plt.figure(figsize=(30,15))
tree.plot_tree(decision_tree, filled=True)
fig.show()
```
